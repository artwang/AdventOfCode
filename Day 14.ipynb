What is the fewest number of seconds that must elapse for the robots to display the Easter egg?

Answer: 
----------------------------
Look for a "region" (adjacent robots) that are more than 100
print the map and look for the tree


```python
##************************************************
## *****   Part II Main Program Start Here   *****
##************************************************
import os
from collections import deque

def get_neighbors(coord):
    x, y = coord
    return [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]

## find groups of adjacent nodes from a list of nodes (might or might not in the same group)
def group_coordinates(coords):
    visited = set()
    groups = []

    for coord in coords:
        if coord not in visited:
            group = []
            queue = deque([coord])
            visited.add(coord)

            while queue:
                current = queue.popleft()
                group.append(current)

                for neighbor in get_neighbors(current):
                    if neighbor in coords and neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)

            if len(group) > 100:
                groups.append(group)
                break

    return groups


## store the intial input into a list
inputData = []
with open('D:\Work\AdventOfCode\Data\Day 14 Data.txt','r') as f:
    for line in f:    # change x to y and y to x
        px = int(line.split(' ')[0].split('=')[1].split(',')[0])
        py = int(line.split(' ')[0].split('=')[1].split(',')[1])
        vx = int(line.split(' ')[1].split('=')[1].split(',')[0])
        vy = int(line.split(' ')[1].split('=')[1].split(',')[1])
        inputData.append((px,py,vx,vy))

# initial variable values
xbound = 101     # 101 wide
ybound = 103     # 103 tall
## prepare 101x103 canvas
#canvas = ("." * ybound + "\n")*xbound

#print(inputData)
#
forRange = (0, 90000, 1)  # (start, end, step)

for move in range(forRange[0],forRange[1],forRange[2]):
    # clear = lambda: os.system('cls')
    # clear()

    lstNextMove = []
    for i in range(len(inputData)):
        (px,py,vx,vy) = inputData[i]
        lstNextMove.append(((px + vx * move) % xbound,(py + vy * move) % ybound))
    lstNextMove = sorted(list(set(lstNextMove)))    # dedup and sort

    cluster = group_coordinates(lstNextMove)
    if len(cluster)>0:
        print(lstNextMove)
        print(cluster)
        print("move with possible Xmas tree = ", move)

        # redraw the map/canvas
        lstCanvas = []
        for i in range(ybound):
            str = list("." * xbound + "\n")
            for x in [x for x,y in lstNextMove if y==i]:
                str[x] = 'O'
            lstCanvas.append("".join(str))
        print("".join(lstCanvas))
        print("")
        break


```

```python
from collections import deque

def get_neighbors(coord):
    x, y = coord
    return ((x+1, y), (x-1, y), (x, y+1), (x, y-1))

def group_coordinates(coords):
    visited = set()
    groups = []

    for coord in coords:
        if coord not in visited:
            group = []
            queue = deque([coord])
            visited.add(coord)

            while queue:
                current = queue.popleft()
                group.append(current)

                for neighbor in get_neighbors(current):
                    if neighbor in coords and neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)

            groups.append(group)

    return groups

# Example usage
coords = [(0, 0), (0, 1), (1, 0), (2, 2), (2, 3), (3, 3)]
groups = group_coordinates(coords)

for idx, group in enumerate(groups):
    print(f"Group {idx + 1}: {group}")
```
