```python
import re

##-----------------------------------------------------------------------
## def to get counts of matched string, forward + backward matched counts
##-----------------------------------------------------------------------
def getCountOfMatch(str2Match):
    count=0
    for i in range(len(str2Match)):
        # forward
        rest = [i.start() for i in re.finditer(find,str2Match[i])]
        count += len(rest)
        # backwards
        rest = [i.start() for i in re.finditer(find,str2Match[i][::-1])]
        count += len(rest)
    print(count)
    return count

##---------------------------------------------------------------------------
## def to get the string list to be searched
##  assemble the input text to be "cut" diagonally from the top left conner
##---------------------------------------------------------------------------
def getDiagonalSearchText(searchList):
    procText=[]
    for i in range(len(searchList[0])):     # cut string according to the top row
        strList=[]
        for j in range(i+1):
            strList.append(searchList[j][i-j])  # cut from top left to the bottom left
        procText.append("".join(strList))

    return procText

##****************************************
## *****   Main Program Start Here   *****
##****************************************

## initialize variables
allText = []
find = 'XMAS'
result = 0

##-------------------------------------------------
## load data file
##-------------------------------------------------
with open('D:\Work\AdventOfCode\Data\Day 04 Data.txt','r') as f:
    for line in f:
        allText.append(line.replace("\n",""))   # to avoid out of index error

##-----------------------------
## for horizontal (#424)
##-----------------------------
result += getCountOfMatch(allText)

##-----------------------------
# for vertical (#441)
##-----------------------------
procText=[]
for i in range(len(allText[0])):
    strList=[]
    for j in range(len(allText)):
        strList.append(allText[j][i])
    procText.append("".join(strList))

result += getCountOfMatch(procText)

##--------------------------------------------------
# for diagonal - top left (#403)
##--------------------------------------------------
procText=[]
altText=allText

procText = getDiagonalSearchText(altText)

result += getCountOfMatch(procText)

##--------------------------------------------------
# for diagonal - bottom right (#404)
##--------------------------------------------------
procText=[]
altText=[]
# rearrange bottom right to the top left
for i in range(len(allText[0])):
    strList=[]
    for j in range(len(allText)):
        strList.append(allText[len(allText)-j-1][len(allText[0])-i-1])
    altText.append(strList)

procText = getDiagonalSearchText(altText)

#   the last string is repeated with top left
del procText[len(procText)-1]

result += getCountOfMatch(procText)

##--------------------------------------------------
# for diagonal - bottom left (#353)
##--------------------------------------------------
procText=[]
# rearrange bottom left to the top left
altText = allText[::-1]

procText = getDiagonalSearchText(altText)

result += getCountOfMatch(procText)

##--------------------------------------------------
# for diagonal - top right (#347)
#   the last string is repeated with top left
##--------------------------------------------------
procText=[]
altText=[]
# rearrange top right to the top left
for i in range(len(allText[0])):
    strList=[]
    for j in range(len(allText)):
        strList.append(allText[j][len(allText[0])-i-1])
    altText.append(strList)

procText = getDiagonalSearchText(altText)

#   the last string is repeated with top left
del procText[len(procText)-1]

result += getCountOfMatch(procText)

#----------------------------------
# display final counts
##----------------------------------
print(f"Total Count: {result}")
```
