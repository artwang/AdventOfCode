```python
## *****************************
## Part II Program
## *****************************
def controller2controllerCode(bCode, eCode):
    if bCode == 'A' and eCode == '^': return "<"    #2a
    if bCode == 'A' and eCode == '>': return "v"    #2d
    if bCode == 'A' and eCode == 'v': return "<v"   #3c
    if bCode == 'A' and eCode == '<': return "v<<"  #4a
    if bCode == '^' and eCode == 'A': return ">"    #2b
    if bCode == '>' and eCode == 'A': return "^"    #2c
    if bCode == 'v' and eCode == 'A': return "^>"   #3g instead of ">^"   #3e
    if bCode == '<' and eCode == 'A': return ">>^"  #4b
    if bCode == '^' and eCode == '>': return "v>A"  #3i instead of ">v"   #3f
    if bCode == '^' and eCode == 'v': return "v"    #2d
    if bCode == '^' and eCode == '<': return "v<"   #3j not interchangable, cannot be "<v"
    if bCode == '>' and eCode == '^': return "<^"   #3b
    if bCode == 'v' and eCode == '^': return "^"    #2c
    if bCode == '<' and eCode == '^': return ">^"   #3e not interchangable, cannot be "^>"
    if bCode == '>' and eCode == 'v': return "<"    #2a
    if bCode == '>' and eCode == '<': return "<<"   #3a
    if bCode == 'v' and eCode == '>': return ">"    #2b
    if bCode == '<' and eCode == '>': return ">>"   #3d
    if bCode == 'v' and eCode == '<': return "<"    #2a
    if bCode == '<' and eCode == 'v': return ">"    #2b
    return ""

##==================================================
## convert the move to symbol
##==================================================
def getMoveSymbol(moves):
    if moves=='A'   : return '1'
    if moves=='<A'  : return '2a'
    if moves=='>A'  : return '2b'
    if moves=='^A'  : return '2c'
    if moves=='vA'  : return '2d'
    if moves=='<<A' : return '3a' #
    if moves=='<^A' : return '3b' # over 3h
    if moves=='<vA' : return '3c' # over 3j
    if moves=='>>A' : return '3d' #
    if moves=='>^A' : return '3e'
    if moves=='>vA' : return '3f'
    if moves=='^>A' : return '3g' # over 3e
    if moves=='^<A' : return '3h'
    if moves=='v>A' : return '3i' # over 3f
    if moves=='v<A' : return '3j'
    if moves=='v<<A': return '4a'
    if moves=='>>^A': return '4b'

##==================================================
## precalculate the next moves - The final optimal order is <, v, ^, >
##==================================================
def getNextMoves(move):
    if move=='1':  return ('1')
    if move=='2a': return ('4a','4b')
    if move=='2b': return ('2d','2c')
    if move=='2c': return ('2a','2b')
    if move=='2d': return ('3c','3g')           #('3c','3e') ('3c','3g') ('3j','3e') ('3j','3g')
    if move=='3a': return ('4a','1','4b')
    if move=='3b': return ('4a','3e','2b')
    if move=='3c': return ('4a','2b','3g')      #('4a','2b','3e') ('4a','2b','3g')
    if move=='3d': return ('2d','1','2c')
    if move=='3e': return ('2d','3b','2b')      #('2d','3b','2b') ('2d','3h','2b')
    if move=='3f': return ('2d','2b','3g')      #('2d','2b','3e') ('2d','2b','3g')
    if move=='3g': return ('2a','3i','2c')      #('2a','3f','2c') ('2a','3i','2c')
    if move=='3h': return ('2a','3j','4b')
    if move=='3i': return ('3c','2b','2c')      #('3c','2b','2c') ('3j','2b','2c')
    if move=='3j': return ('3c','2a','4b')      #('3c','2a','4b') ('3j','2a','4b')
    if move=='4a': return ('3c','2a','1','4b')  #('3c','2a','1','4b') ('3j','2a','1','4b')
    if move=='4b': return ('2d','1','3b','2b')  #('2d','1','3b','2b') ('2d','1','3h','2b')
    # if move=='4c': return ('4a','2b','2a','4b')
    # if move=='4d': return ('2d','3b','3i','2c')

##==================================================
## function to add the count to each occurrence
##==================================================
def countOccurrences(lstMoves):
    lstPads = [('1',0),('2a',0),('2b',0),('2c',0),('2d',0),('3a',0),('3b',0),('3c',0),('3d',0),('3e',0),('3f',0),('3g',0),('3h',0),('3i',0),('3j',0),('4a',0),('4b',0)]
    for move in lstMoves:
        multiplier = move[1]
        for nextmove in getNextMoves(move[0]):
            newPad = (nextmove, [item for item in lstPads if item[0]==nextmove][0][1]+multiplier)
            lstPads = [newPad if item[0]==nextmove else item for item in lstPads]
    nextMoves = [item for item in lstPads if item[1]>0]
    #print(nextMoves)
    return nextMoves

##-------------------------
## input codes
##-------------------------
strCodes = """671A
083A
582A
638A
341A"""
lstKeyCodes = []
for line in strCodes.splitlines():
    lstKeyCodes.append(line.replace("\n",""))

##---------------------------------------------------
##-- create all keypad position
##---------------------------------------------------
keyPad = """789
456
123
#0A"""
lstKeyPad = [((3,1),'0'),((3,2),'A')]
for i in range(1,10):
    lstKeyPad.append(((int((9-i)/3),(i+2)%3),str(i)))
#print(lstKeyPad)

## loop through all codes
complexities = 0
for code in lstKeyCodes:

    ##---------------------------------------------------
    ##-- keypad to 1st controller
    ##---------------------------------------------------
    xb,yb = (3,2)   # always start with 'A'
    sumPathLength = 0
    for k in range(len(code)):
        xe,ye = next((item for item in lstKeyPad if item[1]==code[k]), ((3,2),'A'))[0]
        path = []

        ##---------------------------------------------------
        ##---------------------------------------------------
        up, dn, rt, lt = "","","",""
        if xe-xb > 0:   up = ('v'*(xe-xb))
        if xe-xb < 0:   dn = ('^'*(xb-xe))
        if ye-yb > 0:   rt = ('>'*(ye-yb))
        if ye-yb < 0:   lt = ('<'*(yb-ye))
        path.append(rt+lt+up+dn)
        firstCodes = ''.join(path)
        #print(firstCodes)

        import itertools
        setAllPaths = set(list(itertools.permutations(firstCodes)))
        #print(code[k], len(setAllPaths), setAllPaths)

        ##---------------------------------------------------
        ## loop through all possible path (permutations)
        ##---------------------------------------------------
        lenPath, shortestPath = 0, ""
        for path in setAllPaths:
            #print("path=",path)
            ##---------------------------------------------------
            ## cannot step on (3,0), skip it
            ##---------------------------------------------------
            if ((xb,yb) == (3,1) and ((xe,ye) in [(0,0),(1,0),(2,0)]) and ''.join(path[:1]) == '<') or \
                ((xb,yb) == (3,2) and ((xe,ye) in [(0,0),(1,0),(2,0)]) and ''.join(path[:2]) == '<<') or \
                ((xb,yb) == (0,0) and ((xe,ye) in [(3,1),(3,2)]) and ''.join(path[:3]) == 'vvv') or \
                ((xb,yb) == (1,0) and ((xe,ye) in [(3,1),(3,2)]) and ''.join(path[:2]) == 'vv') or \
                ((xb,yb) == (2,0) and ((xe,ye) in [(3,1),(3,2)]) and ''.join(path[:1]) == 'v'):
                #print("skip",path,"for",xb,yb,xe,ye)
                continue

            firstCodes = ''.join(path)+'A'
            #print("firstCodes:", firstCodes)

            ## 25 directional keypads for robots
            robotCodes = firstCodes

            ##---------------------------------------------------
            ##-- 1st controller to 2nd controller
            ##  apply the 1st controller code, so the moves will be limited to 3 (4 including A at the end)
            ##---------------------------------------------------
            bCode = 'A'
            lstCode=[]
            for i in range(len(firstCodes)):
                eCode = firstCodes[i]
                lstCode.append(controller2controllerCode(bCode, eCode)+'A')
                bCode = eCode
            #print(lstCode)

            ##---------------------------------------------------
            ## loop the rest iterations with simplified counts
            ##---------------------------------------------------
            lstMoves = [(getMoveSymbol(item),1) for item in lstCode]
            #print(lstMoves)
            for m in range(24):
                lstMoves = countOccurrences(lstMoves)
                TotalMoves = 0
                for i in range(len(lstMoves)):
                    TotalMoves += int(lstMoves[i][0][:1])*lstMoves[i][1]
            #print("firstCodes=",firstCodes, ", Length=", TotalMoves)

            ##---------------------------------------------------
            ## compare the record the shortest path
            ##---------------------------------------------------
            if lenPath==0 or TotalMoves<lenPath:
                lenPath, shortestPath = TotalMoves, firstCodes

        print("code:",code[k], "keypad path:", shortestPath, ',Len = ', lenPath)
        sumPathLength += lenPath

        ## make the end node to be the beginning one for next round
        xb,yb = xe,ye

    print("Sum of Path Length = ", sumPathLength)
    complexities += int(code.replace('A',''))*sumPathLength

print("Total Complexities = ", complexities)


```
