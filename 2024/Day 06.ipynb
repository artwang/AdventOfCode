```python
##********************
# Part I Test sample
##********************
strMap = """************
*....#.....*
*.........#*
*..........*
*..#.......*
*.......#..*
*..........*
*.#..^.....*
*........#.*
*#.........*
*......#...*
************"""
#print(strMap)
lstMap = []
for line in strMap.splitlines():
    lstMap.append(line)

gameOver = False
currentDir = ''
currentPos = ()
lstTrace = []
result = 1  # count the starting point as 1

## find the starting point and its direction
for x in range(len(lstMap)):
    for y in range(len(lstMap[0])):
        if lstMap[x][y] == '^' or lstMap[x][y] == 'v' or lstMap[x][y] == '<' or lstMap[x][y] == '>' :
            currentPos = (x,y)
            currentDir = lstMap[x][y]
            lstTrace.append(currentPos)
            print(f"starting point: ({x},{y}) face {'up' if lstMap[x][y] == '^' else ('down' if lstMap[x][y] == 'v' else ('left' if lstMap[x][y] == '<' else 'right'))}")
            break
    if currentDir != '':
        break

## marching
while not gameOver:
    (x,y) = currentPos
    nextPos = (x-1,y) if currentDir == '^' else ((x+1,y) if currentDir == 'v' else ((x,y-1) if currentDir == '<' else (x,y+1)))
    nextSymbol = lstMap[nextPos[0]][nextPos[1]]
    print(f"nextSymbol = {nextSymbol}")

    ## next step - go up - check the next cell
    if nextSymbol == '*' :  # stop
        gameOver = True
        break
    elif nextSymbol == '#' :  # obstacle - turn right
        currentDir = '>' if currentDir == '^' else ('<' if currentDir == 'v' else ('^' if currentDir == '<' else 'v'))
        print(f"currentDir = {currentDir}")
    else:   # else nextSymbo == '.' :  # step in (continue)
        currentPos = (x-1,y) if currentDir == '^' else ((x+1,y) if currentDir == 'v' else ((x,y-1) if currentDir == '<' else (x,y+1)))
        lstTrace.append(currentPos)     ## increment only when moving into next .
        print(f"currentPos = {currentPos}")

## dedup lstTrace
lstTrace = list(set(lstTrace))

print(f"Total steps = {len(lstTrace)}")


```

It doesn't really matter what you choose to use as an obstacle so long as you and The Historians can put it into position without the guard noticing. The important thing is having enough options that you can find one that minimizes time paradoxes, and in this example, there are 6 different positions you could choose.

You need to get the guard stuck in a loop by adding a single new obstruction. How many different positions could you choose for this obstruction?

Answer: 
--------------------------------------------
1. random pick a cell (besides it occupied by an obstacle or the starting point) and change it to '#'
2. start traversing the cell and record the visited cells with coordinate and direction
3. after moving to the next cell or making a direction change (to the right), check if the cell has been visited before with the same direction
    a. if yes, the picked cell is an option
    b. if reach the border (*), the picked cell is not an option
    c. else move on to the next move

* Instead of running/testing all cells (10x10, or 131x131), only test on those cells will be visited (41, or 4982) before getting out of bound

```python
##**************************
# Part II Test sample map
##**************************
strMap = """************
*....#.....*
*.........#*
*..........*
*..#.......*
*.......#..*
*..........*
*.#..^.....*
*........#.*
*#.........*
*......#...*
************"""
#print(strMap)
lstMap = []
for line in strMap.splitlines():
    lstMap.append(line)

gameOver = False
startPos = ()       ## use 3-element tupple (x,y,dir)
currentPos = ()
lstTrace = []
lstObstacle = []
result = 1  # count the starting point as 1

##-----------------------------------------------
## find the starting point and its direction
##-----------------------------------------------
for x in range(len(lstMap)):
    for y in range(len(lstMap[0])):
        if lstMap[x][y] == '^' or lstMap[x][y] == 'v' or lstMap[x][y] == '<' or lstMap[x][y] == '>' :
            startPos = (x,y,lstMap[x][y])
            print(f"starting point: ({x},{y},{lstMap[x][y]}) face {'up' if lstMap[x][y] == '^' else ('down' if lstMap[x][y] == 'v' else ('left' if lstMap[x][y] == '<' else 'right'))}")
            break
    if startPos != ():
        break

##-------------------------------------------------------------------------------------
## loop through all cells as candidate of obstacle for the guard to stuck in the loop
##-------------------------------------------------------------------------------------
for x in range(len(lstMap)):
    for y in range(len(lstMap[0])):
        ## skip border cells and starting point (only take the cell is a dot)
        if lstMap[x][y] == '.' :

            ## copy to a new map, so not need to revert the change for the next loop
            lstNewMap = lstMap.copy()

            ## assign obstacle
            obstacle = (x,y)
            lstNewMap[x] = lstNewMap[x][:y]+"#"+lstNewMap[x][y+1:]  # string is immutable, so it needs to be replaced as a whole string
            #print(f"obstacle = ({x},{y})")

            ## initialize variables
            gameOver = False
            bingo = False
            lstTrace = []
            lstTrace.append(startPos)
            currentPos = startPos

            ##-------------------
            ## marching
            ##-------------------
            while not gameOver:
                (a,b,currentDir) = currentPos
                nextPos = (a-1,b) if currentDir == '^' else ((a+1,b) if currentDir == 'v' else ((a,b-1) if currentDir == '<' else (a,b+1)))
                nextSymbol = lstNewMap[nextPos[0]][nextPos[1]]
                #print(f"nextSymbol = {nextSymbol}")

                ## next step - go up - check the next cell
                if nextSymbol == '*' :  # stop
                    gameOver = True
                    break
                elif nextSymbol == '#' :  # obstacle - turn right
                    currentPos = (a,b,'>' if currentDir == '^' else ('<' if currentDir == 'v' else ('^' if currentDir == '<' else 'v')))
                    #print(f"currentDir = {currentPos[2]}")
                else:   # else nextSymbo == '.' :  # step in (continue)
                    currentPos = (a-1,b,currentDir) if currentDir == '^' else ((a+1,b,currentDir) if currentDir == 'v' else ((a,b-1,currentDir) if currentDir == '<' else (a,b+1,currentDir)))

                ## check if the same pos is visted again
                if currentPos in lstTrace:
                    #print(f"repeated pos: ({currentPos[0]},{currentPos[1]},{currentPos[2]})")
                    bingo = True
                    break

                lstTrace.append(currentPos)     ## increment only when moving into next .
                #print(f"currentPos = ({currentPos[0]},{currentPos[1]},{currentPos[2]})")

            ##
            if bingo :
                lstObstacle.append(obstacle)
                print(f"bingo = ({obstacle[0]},{obstacle[1]})")

## print result
print(f"Total Obstacle Options to Make Loop = {len(lstObstacle)}")
```
