To begin, get your puzzle input.

Answer: 
-------------------------------------------------  
1. get all network computer pairs 
2. get all distinct computers  
3. get all computers started with 't'  
4. get all combinations with 3 elements  
5. filter those combinations with at least one computer with leading 't' character  
6. check anc count the filtered computer group (3 computers) have network connections for each other  
7. it took 18.2 seconds to run, probably not going to fly with Part 2.  


```python
##*****************************
# Part I Test Program
##*****************************
network="""kh-tc
qp-kh
de-cg
ka-co
yn-aq
qp-ub
cg-tb
vc-aq
tb-ka
wh-tc
yn-cg
kh-ub
ta-co
de-co
tc-td
tb-wq
wh-td
ta-ka
td-qp
aq-cg
wq-ub
ub-vc
de-ta
wq-aq
wq-vc
wh-yn
ka-de
kh-ta
co-tc
wh-qp
tb-vc
td-yn"""

## get all computers
lstComputers=[]
setNetwork=set()
for line in network.splitlines():
    lstComputers.append(line.replace("\n","").split('-')[0])
    lstComputers.append(line.replace("\n","").split('-')[1])
    setNetwork.add(tuple(line.replace("\n","").split('-')))
setComputers = set(lstComputers)
lstComputers = list(setComputers)
setTComputers = set([item for item in lstComputers if item[:1]=='t'])
print(setNetwork)
print(lstComputers)
print(setComputers)
print(setTComputers)

import itertools
lstGroup = list(itertools.combinations(setComputers,3))
lstTGroup = [item for item in lstGroup if item[0][0]=='t' or item[1][0]=='t' or item[2][0]=='t']
print(len(lstGroup), lstGroup)
print(len(lstTGroup), lstTGroup)
lstInNetwork = [item for item in lstTGroup if   ((item[0],item[1]) in setNetwork or (item[1],item[0]) in setNetwork) \
                                            and ((item[1],item[2]) in setNetwork or (item[2],item[1]) in setNetwork) \
                                            and ((item[0],item[2]) in setNetwork or (item[2],item[0]) in setNetwork)]

print(len(lstInNetwork), lstInNetwork)
```

--- Part Two ---
There are still way too many results to go through them all. You'll have to find the LAN party another way and go there yourself.

Since it doesn't seem like any employees are around, you figure they must all be at the LAN party. If that's true, the LAN party will be the largest set of computers that are all connected to each other. That is, for each computer at the LAN party, that computer will have a connection to every other computer at the LAN party.

In the above example, the largest set of computers that are all connected to each other is made up of co, de, ka, and ta. Each computer in this set has a connection to every other computer in the set:

The LAN party posters say that the password to get into the LAN party is the name of every computer at the LAN party, sorted alphabetically, then joined together with commas. (The people running the LAN party are clearly a bunch of nerds.) In this example, the password would be co,de,ka,ta.

What is the password to get into the LAN party?

Answer:  
---------------------------------------  
a connection can join to more than one group  
how to find a connection group:  
1. sort the connection pair among the pairs and within the connection  
2. process all connections with a designated computer one at a time from alphabetical order - no need to include the previous computers that were processed (that's why processing them in alphabetical order)   
3. get all computers connected with this designated computer  
4. generate expected combinations from this list of computers  
5. check computer with the most missing connections and remove it from the list from #3  
6. process #4 & #5 unitl no more missing computer
7. record the list of computer as a network group and compare to the previous groups
8. shortcut the process to check if the number of computers in the processing group is less or equal to the previous max one. No need to process them further.  

```python
##*****************************
# Part II Test Program
##*****************************
network="""kh-tc
qp-kh
de-cg
ka-co
yn-aq
qp-ub
cg-tb
vc-aq
tb-ka
wh-tc
yn-cg
kh-ub
ta-co
de-co
tc-td
tb-wq
wh-td
ta-ka
td-qp
aq-cg
wq-ub
ub-vc
de-ta
wq-aq
wq-vc
wh-yn
ka-de
kh-ta
co-tc
wh-qp
tb-vc
td-yn"""

## get all computers
lstComputers=[]
setNetwork=set()
for line in network.splitlines():
    a,b = line.replace("\n","").split('-')[0],line.replace("\n","").split('-')[1]
    lstComputers.append(a)
    lstComputers.append(b)
    setNetwork.add((a if a<b else b, b if a<b else a))
setComputers = set(lstComputers)
lstComputers = sorted(list(setComputers))
# print("setNetwork",len(setNetwork), setNetwork)
# print(len(setNetwork), sorted(list(setNetwork)))
# print("setComputers",len(setComputers), setComputers)
# print("lstComputers", len(lstComputers),lstComputers)

maxGroup, lstMaxGroup = 0, []

for i in range(len(lstComputers)):
    # get computer connect/start with ab
    computer=lstComputers[i]
    lstABNetwork = [item for item in setNetwork if item[0]==computer]
    print("lstABNetwork",len(lstABNetwork), lstABNetwork)
    lstABComputer = sorted([item[1] for item in lstABNetwork])
    lstABComputer.insert(0,computer)
    print("lstABComputer", len(lstABComputer), lstABComputer)

    ## if the total len is already smaller than the recorded maxGroup, no need to process further
    if len(lstABComputer) <= maxGroup:
        continue

    ## loop until the max group is detected
    while True:
        ## generate combinations of target computers
        lstABCombined = sorted(list(itertools.combinations(lstABComputer,2)))
        #print("lstABCombined",len(lstABCombined), lstABCombined)
        # lstABCombinedInNetwork = sorted([item for item in lstABCombined if item in setNetwork])
        # print("lstABCombinedInNetwork",len(lstABCombinedInNetwork), lstABCombinedInNetwork)

        ## check connections not in the network
        lstABCombinedNotInNetwork = sorted([item for item in lstABCombined if item not in setNetwork])
        print("lstABCombinedNotInNetwork",len(lstABCombinedNotInNetwork), lstABCombinedNotInNetwork)

        if lstABCombinedNotInNetwork==[]:
            maxGroup, lstMaxGroup = len(lstABComputer), lstABComputer
            print("interim maxGroup, lstMaxGroup", maxGroup, lstMaxGroup)
            break

        ## find the next computer (with most missing connections) to remove to make a connection group
        lstABComputerNotInNetwork = list(itertools.chain(*lstABCombinedNotInNetwork))
        dicABComputerNotInNetwork = Counter(lstABComputerNotInNetwork)
        print(dicABComputerNotInNetwork)
        TobeRemovedComputer = max(dicABComputerNotInNetwork, key=dicABComputerNotInNetwork.get)
        print("Next to remove", TobeRemovedComputer)
        lstABComputer.remove(TobeRemovedComputer)

        ## if the total len is already smaller than the recorded maxGroup, no need to process further
        if len(lstABComputer) <= maxGroup:
            break

print("maxGroup, lstMaxGroup", maxGroup, lstMaxGroup)
print("password=",','.join(lstMaxGroup))


```
