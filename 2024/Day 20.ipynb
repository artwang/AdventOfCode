--- Day 20: Race Condition ---
The Historians are quite pixelated again. This time, a massive, black building looms over you - you're right outside the CPU!

While The Historians get to work, a nearby program sees that you're idle and challenges you to a race. Apparently, you've arrived just in time for the frequently-held race condition festival!

The race takes place on a particularly long and twisting code path; programs compete to see who can finish in the fewest picoseconds. The winner even gets their very own mutex!

They hand you a map of the racetrack (your puzzle input). For example:

###############

#.#.#.#.#.###.#

#...#...#...###
###############
The map consists of track (.) - including the start (S) and end (E) positions (both of which also count as track) - and walls (#).

When a program runs through the racetrack, it starts at the start position. Then, it is allowed to move up, down, left, or right; each such move takes 1 picosecond. The goal is to reach the end position as quickly as possible. In this example racetrack, the fastest time is 84 picoseconds.

Because there is only a single path from the start to the end and the programs all go the same speed, the races used to be pretty boring. To make things more interesting, they introduced a new rule to the races: programs are allowed to cheat.

The rules for cheating are very strict. Exactly once during a race, a program may disable collision for up to 2 picoseconds. This allows the program to pass through walls as if they were regular track. At the end of the cheat, the program must be back on normal track again; otherwise, it will receive a segmentation fault and get disqualified.

So, a program could complete the course in 72 picoseconds (saving 12 picoseconds) by cheating for the two moves marked 1 and 2:

Each cheat has a distinct start position (the position where the cheat is activated, just before the first move that is allowed to go through walls) and end position; cheats are uniquely identified by their start position and end position.

```python
##*****************************
# Part I Test Program
##*****************************
strMap = """###############
#...#...#.....#
#.#.#.#.#.###.#
#S#...#.#.#...#
#######.#.#.###
#######.#.#...#
#######.#.###.#
###..E#...#...#
###.#######.###
#...###...#...#
#.#####.#.###.#
#.#...#.#.#...#
#.#.#.#.#.#.###
#...#...#...###
###############"""

## get Map into a list
lstMap = []
for line in strMap.splitlines():
    lstMap.append(line.replace("\n",""))
#print(lstMap)

xbound = len(lstMap)
ybound = len(lstMap[0])

## find S and E
startX, startY, endX, endY = 0, 0, 0, 0
findThem = 0
for i in range(len(lstMap)):
    for j in range(len(lstMap[0])):
        if lstMap[i][j] == 'S':
            startX, startY, findThem = i, j, findThem+1
        if lstMap[i][j] == 'E':
            endX, endY, findThem = i, j, findThem+1

        if findThem == 2: break
    if findThem == 2: break

print("startX, startY, endX, endY = ", startX, startY, endX, endY)

## Run the path and count the picoseconds.
##  Record the picoseconds along with each passable track position (.)
isEnd = False
setVisited = set((startX,startY))
lstVisited = [((startX,startY),0)]
count = 1
x, y = startX, startY
while not isEnd and count < xbound*ybound:
    if lstMap[x-1][y]!='#' and (x-1,y) not in setVisited:
        lstVisited.append(((x-1,y),count))
        setVisited.add((x-1,y))
        x, y, count = x-1, y, count+1
    elif lstMap[x][y-1]!='#' and (x,y-1) not in setVisited:
        lstVisited.append(((x,y-1),count))
        setVisited.add((x,y-1))
        x, y, count = x, y-1, count+1
    elif lstMap[x+1][y]!='#' and (x+1,y) not in setVisited:
        lstVisited.append(((x+1,y),count))
        setVisited.add((x+1,y))
        x, y, count = x+1, y, count+1
    elif lstMap[x][y+1]!='#' and (x,y+1) not in setVisited:
        lstVisited.append(((x,y+1),count))
        setVisited.add((x,y+1))
        x, y, count = x, y+1, count+1
    if lstMap[x][y]=='E':
        isEnd = True
print(lstVisited)

## Find all the cheats (horizontal and vertical) those has pattern of .#.
lstHCheats = []
lstVCheats = []
for i in range(1,len(lstMap)-1):
    for j in range(1,len(lstMap[0])-1):
        if lstMap[i][j]=='#' and lstMap[i-1][j]!='#' and lstMap[i+1][j]!='#':
            lstVCheats.append((i,j))
        if lstMap[i][j]=='#' and lstMap[i][j-1]!='#' and lstMap[i][j+1]!='#':
            lstHCheats.append((i,j))
print(lstHCheat)
print(lstVCheat)

## Run those cheats one at a time (Hm..., there are rooms to improve) - calculate the saving by substract the picoseconds before and after passing the wall to cheat.
cutoff = 10
bigCheatCount = 0
for hCheat in lstHCheats:
    leftPicoSec = (item for item in lstVisited if item[0]==(hCheat[0],hCheat[1]-1)),((),0))[1]
    rightPicoSec = next((item for item in lstVisited if item[0]==(hCheat[0],hCheat[1]+1)),((),0))[1]
    print(hCheat, leftPicoSec, rightPicoSec, abs(leftPicoSec-rightPicoSec)-2)
    if abs(leftPicoSec-rightPicoSec)-2 >= cutoff:
        bigCheatCount += 1


for vCheat in lstVCheats:
    topPicoSec = next((item for item in lstVisited if item[0]==(vCheat[0]-1,vCheat[1])),((),0))[1]
    downPicoSec = next((item for item in lstVisited if item[0]==(vCheat[0]+1,vCheat[1])),((),0))[1]
    print(vCheat, topPicoSec, downPicoSec, abs(topPicoSec-downPicoSec)-2)
    if abs(topPicoSec-downPicoSec)-2 >= cutoff:
        bigCheatCount += 1


## Count the cheats that save at least 100 picoseconds
print("Total Big Cheat Count=", bigCheatCount)
```
